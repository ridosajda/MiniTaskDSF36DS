```python
# deteksi data null, nan, na, dan duplikasi kemudian hapus

import pandas as pd

df = pd.read_csv('seattle-weather.csv')

# Deteksi data null, NaN, dan NA
print("Jumlah data null, NaN, dan NA:")
print(df.isnull().sum())

# Deteksi duplikasi
print("\nJumlah data duplikat:")
print(df.duplicated().sum())

# Hapus data null, NaN, dan NA
df.dropna(inplace=True)

# Hapus data duplikat
df.drop_duplicates(inplace=True)

print("\nData setelah dihapus:")
df
```

```python
df.info()
```

```python
# code label encoder untuk merubah menjadi angka kolom date dan weather

from sklearn.preprocessing import LabelEncoder

# Inisialisasi LabelEncoder
le = LabelEncoder()

# Ubah kolom 'date' menjadi angka
df['date_encoded'] = le.fit_transform(df['date'])

# Ubah kolom 'weather' menjadi angka
df['weather_encoded'] = le.fit_transform(df['weather'])

print(df[['date', 'date_encoded', 'weather', 'weather_encoded']])
```

```python
# tampilkan kolom weather dan weather_encoded tetapi tidak ada duplikasi

print(df[['weather', 'weather_encoded']].drop_duplicates())
```

```python
df.info()
```

```python
import numpy as np

x_train = np.array(df[['precipitation', 'temp_max', 'temp_min', 'wind']]) # konversi data menjadi np.array dan ditampung ke variable x_train sebagai sekumpulan features untuk training set
y_train = np.array(df['weather_encoded'])                                 # konversi data menjadi np.array dan ditampung ke variable y_train sebagai sekumpulan nilai target untuk training set

print(f'x_train:\n{x_train}\n')
print(f'y_train: {y_train}')
```

```python
from sklearn.neighbors import KNeighborsClassifier

K = 6
model = KNeighborsClassifier(n_neighbors = K)
model.fit(x_train, y_train)
```

```python
# melakukan prediksi cuaca dengan precipitation, temp_max, temp_min, dan wind
precipitation = 1.9
temp_max = 15.5
temp_min = 0.1
wind = 2.7
x_new = np.array([precipitation, temp_max, temp_min, wind])
x_new
```

```python
# prediksi cuaca dengan model KNeighborsClassifier yang sudah di training
y_new = model.predict([x_new])
y_new
```

```python
# panggil kembali label encoder untuk mengingatkan nomer 4 itu cuaca apa
le.inverse_transform(y_new)
```

```python
# bentuk data menjadi np.array dan tampung ke varialble misterius
misterius = np.array([precipitation, temp_max, temp_min, wind])
misterius
```

```python
#tampilkan sekumpulan nilai features dari training set
x_train
```

```python
#kalkulasi jarak antara data file misterius dengan data file lain di dalam data file x_train
from scipy.spatial.distance import euclidean

data_jarak = [euclidean(misterius, d) for d in x_train]
data_jarak
```

```python
# membentuk kolom baru dengan nama jarak yang datanya diambil dari variable data_jarak dan di urutkan dar nilai terkecil hingga nilai terbesar
df['jarak'] = data_jarak
df.sort_values('jarak')
```

```python
#testing set
x_test = np.array([[1.9, 15.5, 0.1, 2.7], [3.7, 10.7, 2.0, 3.0], [1.7, 14.7, 0.0, 5.0], [34.7, 9.7, -1.0, 6.0]]) # menyiapkan features untuk target
y_test = le.transform(['sun', 'fog', 'sun', 'sun'])                                                              # menyiapkan target yang sudah di ubah menjadi numeric biner dengan label encoder

print(f'x_test:\n{x_test}\n')
print(f'y_test: {y_test}')
```

```python
#prediksi terhadap testing set
y_pred = model.predict(x_test)
y_pred
```

```python
#akurasi
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, y_pred)        # panggil accuracy score dan menyertakan 2 parameter yaitu y_test yang merupakn target dari testing set, dan y_pred merupakan hasil prediksi hasil dari testing set
print(f'Akurasi: {acc}')
```

```python
#presisi merupakan proporsi dari testing set yang diprediksi sebagai positif oleh model dan benar benar positif
from sklearn.metrics import precision_score

prec = precision_score(y_test, y_pred, average='weighted')     # panggil precision score dan menyertakan 2 parameter yaitu y_test yang merupakn target dari testing set, dan y_pred merupakan hasil prediksi hasil dari testing set. Lalu juga memanggil average "weighted" yang digunakan untuk menghitung rata-rata dari sekumpulan data.
print(f'Presisi: {prec}')
```

```python
#recall adalah proporsi dari testing set yang benar benar positif, yang diprediksi sebagai positif oleh model
from sklearn.metrics import recall_score

rec = recall_score(y_test, y_pred, average='weighted')    # panggil recall score dan menyertakan 2 parameter yaitu y_test yang merupakn target dari testing set, dan y_pred merupakan hasil prediksi hasil dari testing set. Lalu juga memanggil average "weighted" yang digunakan untuk menghitung rata-rata dari sekumpulan data.
print(f'Recall: {rec}')
```

```python
#F1 score adalah nilai rata rata harmonic dari precision dan recall
from sklearn.metrics import f1_score

f1 = f1_score(y_test, y_pred, average='weighted')     # panggil f1 score dan menyertakan 2 parameter yaitu y_test yang merupakn target dari testing set, dan y_pred merupakan hasil prediksi hasil dari testing set. Lalu juga memanggil average "weighted" yang digunakan untuk menghitung rata-rata dari sekumpulan data
print(f'F1 score: {f1}')
```

```python
#classification report
from sklearn.metrics import classification_report
cls_report = classification_report(y_test, y_pred)        # panggil classification report dan menyertakan 2 parameter yaitu y_test yang merupakn target dari testing set, dan y_pred merupakan hasil prediksi hasil dari testing set.Lalu ditampung ke variable cls_report
print(f'Classification Report:\n{cls_report}')
```

```python
#matthews correlation coefficient (mcc) adalah alternatif untuk f1 score sebagai keperluan pengukuran performa dari binary classifiers
from sklearn.metrics import matthews_corrcoef
mcc = matthews_corrcoef(y_test, y_pred)           # panggil matthews corrcoef dan menyertakan 2 parameter yaitu y_test yang merupakn target dari testing set, dan y_pred merupakan hasil prediksi hasil dari testing set.Lalu ditampung ke variable mcc
print(f'MCC: {mcc}')
```
